## 统计工具 

```python
 #统计指标1 
    
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,100,size = (20,3)),
index = list('ABCDEFHIJKLMNOPQRSTU'), columns=['Python','Tensorflow','Keras'])

df.iloc[1,2] = None

s1 = df.count(axis = 1) # 调整横纵轴计数 不会计算空值
s2 = df.max(axis = 0 ) # 纵向最值
s3 = df.max(axis = 1 ) # 横向最值
s4 = df.median() # 中位数 默认纵向
s5 = df.sum() # 求和 默认纵向
s6 = df.mean(axis = 1 ) # 均值 默认纵向
s7 = df.quantile(q = [0.2,0.4,0.8]) # 中位数 默认纵向
s8 = df.describe() # 汇总统计

pprint('s',8)
```

```python
 #统计指标2 

sx1 = df["Python"].value_counts()
sx2 = df['Keras'].unique()
sx3 = df.cumsum() # 遇到空值累计的时候会跳过去
sx4 = df.cumprod() # 遇到空值累计的时候会跳过去
sx5 = df.std()
sx6 = df.var()
sx7 = df.cummin() # 类似于滑动窗口， 比较区域是前边所有加上当前 不看之后
sx8 = df.cummax() # 类似于滑动窗口， 比较区域是前边所有加上当前 不看之后
sx9 = df.diff() # 差分就是后边的一个减去前边的一个 
sx10 = df.pct_change() # 前后环比
print(df) 
pprint('sx',10)
```

```python
# 统计指标3 高级

sh1 = df.cov() # 协方差矩阵
sh2 = df['Python'].cov(df['Keras']) # 协方差
sh3 = df.corr() # 属性相关系数
sh4 = df.corrwith(df['Tensorflow']) # 所有属性 和 指定属性做相关系数

pprint('sh',4)
```

# 案例 拉勾网数据分析师招聘数据预处理

```python
import pandas as pd 
import numpy as np 

job = pd.read_csv('./lagou2020.csv')
job.drop_duplicates(inplace = True) # 去重 inplace true 是 在原对象上修改

job.shape
```

```python
job.head(5)
```

```python
cond = job['positionName'].str.contains("数据分析") # 把 职位名称包含“数据名称”作为筛选条件
job = job[cond] # 筛选
job.reset_index(inplace = True) # 修改原DF
job.shape
```

```python
# 薪资处理 提取两端数值 取均值
job['salary'] = job['salary'].str.lower()\
                .str.extract(r'(\d+)[k]-(\d+)k')\
                .applymap(lambda x:int(x))\
                .mean(axis = 1 )
job.head(3)

```

```python
job['job_detail'] = job['job_detail'].str.lower().fillna('')
job["Python"] = job['job_detail'].map(lambda x:1 if ('Python' in x ) else 0)
job["SQL"] = job['job_detail'].map(lambda x:1 if ('sql' in x  or 'hive' in x) else 0)
job["Tableau"] = job['job_detail'].map(lambda x:1 if ('tableau' in x ) else 0)
job["Excel"] = job['job_detail'].map(lambda x:1 if ('excel' in x ) else 0)
job["SPSS/SAS"] = job['job_detail'].map(lambda x:1 if ('spss' in x or 'sas' in x ) else 0)

job.head(20)
```

```python
def clean_industry(industry) : 
    industry = industry.split(',')
    if industry[0] == '移动互联网' and len(industry) > 1 :
        return industry[1]
    else:
        return industry[0]
    
job['industryField'] = job.industryField.map(clean_industry)

job.head(20)
```
