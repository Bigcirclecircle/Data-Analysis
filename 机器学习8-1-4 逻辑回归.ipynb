```python
import numpy as np
import pandas as pd 
from sklearn.datasets import load_breast_cancer 
import matplotlib.pyplot as plt 
```

```python
X  = load_breast_cancer().data
Y  =  load_breast_cancer().target

pd.DataFrame(X) # çœ‹ä¸€ä¸‹æ•°æ®é›† é‡çº²å·®å¼‚å¤§ è€ƒè™‘åšæ ‡å‡†åŒ– 
```

```python
# å»ºæ¨¡

from sklearn.linear_model import LogisticRegression as LR 
from sklearn.model_selection import train_test_split

# æ€Žä¹ˆçœ‹æ–¹æ³•çš„å‚æ•° ï¼Ÿ  ï¼Ÿ ï¼Ÿ 

help(LR)

# å²­å›žå½’å’Œ LASSO åœ¨çº¿æ€§å›žå½’é‡Œ è§£å†³å…±çº¿æ€§çš„é—®é¢˜ 
# åœ¨è¿™é‡Œè§£å†³çš„æ˜¯è¿‡æ‹Ÿåˆé—®é¢˜  

# L2 æ­£åˆ™åŒ– å¯¹åº”å²­å›žå½’
# L1 æ­£åˆ™åŒ– å¯¹åº”LASSOå›žå½’  é»˜è®¤æ˜¯L2 


```

```python
lr2 = LR (penalty = 'l2',solver = 'liblinear',
   C = 0.5 ,
   max_iter = 1000).fit(X,Y)

lr2.score(X,Y) # ç”¨l1 å’Œ l2 å·®ä¸å¤š l1ç¨å¾®é«˜ä¸€ç‚¹
```

```python
# å’Œ l1 çš„åŒºåˆ«çš„æ˜¯ å½±å“å¾ˆå°çš„é‡ä¹Ÿä¼šè¢«ä¿ç•™ä¸‹æ¥
lr2.coef_

# ç”¨è¿™ä¹ˆå¤šç‰¹å¾æ¥è®­ç»ƒæ¨¡åž‹è‚¯å®šå¤æ‚  ç‰¹å¾å¤ªå¤šä¼šè¿‡æ‹Ÿåˆ  ç‰¹å¾è¶Šå°‘æ³›åŒ–èƒ½åŠ›è¶Šå¼º  ä½†ä¹Ÿä¸èƒ½å¤ªå°‘ é‚£æ ·ä¼šæ¬ æ‹Ÿåˆ 
```

```python
# ç»˜åˆ¶å­¦ä¹ æ›²çº¿ åŽ»çœ‹ l1 å¥½ è¿˜æ˜¯l2 å¥½ 

Xtrain,Xtest,Ytrain,Ytest  = train_test_split(X,Y,test_size = 0.3 , random_state = 420 )
```

```python
# æŸ¥çœ‹C åœ¨L1 L2 ä¸‹ è®­ç»ƒé›†å’Œæµ‹è¯•é›†çš„è¡¨çŽ° 

l1 = []
l2 = [] 

l1test = [] 
l2test = []


for i in np.linspace(0.05,1,20):
    # å®žä¾‹åŒ–æˆ‘çš„æ¨¡åž‹ å¹¶ä¸”è®­ç»ƒ 
    lrl1 = LR(penalty = 'l1',
      solver = 'liblinear',
      C = i,max_iter = 1000).fit(Xtrain,Ytrain)
    
    lrl2 = LR(penalty = 'l2',
      solver = 'liblinear',
      C = i,max_iter = 1000).fit(Xtrain,Ytrain)
    
    l1.append(lrl1.score(Xtrain,Ytrain))
    l2.append(lrl2.score(Xtrain,Ytrain))
    
    l1test.append(lrl1.score(Xtest,Ytest))
    l2test.append(lrl2.score(Xtest,Ytest))
```

```python
graph = [l1,l2,l1test,l2test]
color = ['green','black','lightgreen','gray']
label = ['L1','L2','L1test','L2test']

plt.figure(figsize = (6,6))

for i in range(len(graph)):
    plt.plot(np.linspace(0.05,1,20),graph[i],color[i],label = label[i])
    
plt.legend(loc = 4)
plt.show()


# å­¦ä¹ æ›²çº¿æ­£å¸¸æ˜¯ä¸‹æ»‘çš„ 

# çŽ°åœ¨çš„å›¾ä¸Šå‡  å› ä¸ºCæ˜¯ä¹˜æ³•çš„é€†  Cè¶Šå¤§çš„æ—¶å€™æƒ©ç½šåŠ›åº¦è¶Šå°


# ä¸Šè¾¹æ˜¯è®­ç»ƒé›† ä¸‹è¾¹æ˜¯æµ‹è¯•é›†

# æƒ©ç½šåŠ›åº¦å¤§æ˜¯å•¥æ„æ€ ï¼š 
```

```python
# coding=utf-8
import random
import matplotlib.pyplot as plt
import numpy as np

x, y = [], []
x_test1, x_test2, x_test3 = [], [], []

# éšæœºç”Ÿæˆ3ç§ä¸åŒåˆ†ç±»çš„ç‚¹ï¼Œåˆ†åˆ«æ‰“ä¸Šæ ‡ç­¾å­˜åœ¨yä¸­
for i in range(0, 20):
    x1 = random.random()
    x2 = random.random()
    if x1 + x2 < 1:
        x.append([x1, x2, 1])
        x_test1.append([x1, x2])
        y.append([1, 0, 0])

    x.append([x1 * 2, x2 + 1, 1])
    x_test2.append([x1 * 2, x2 + 1])
    y.append([0, 1, 0])

    if x1 > x2:
        x.append([x1 + 1, x2, 1])
        x_test3.append([x1 + 1, x2])
        y.append([0, 0, 1])

# å°†listè½¬æ¢ä¸ºnumpy array
x = np.array(x)
y = np.array(y)

# å­¦ä¹ çŽ‡
lr = 0.5

w = np.ones((3, 3))
m = len(x)
for i in range(0, 1000):
    w_gard = 0
    for j in range(0, m):
        # sigmoidå‡½æ•°åˆ†ç±»ï¼Œè¿™é‡Œç”¨åˆ°äº†çŸ©é˜µä¹˜æ³•
        w_gard += -x[j] * np.atleast_2d(1 / (1 + np.exp(-np.matmul(w, x[j]))) - y[j]).T
    w += lr * w_gard / m

# ç”»ç‚¹
plt.plot(np.array(x_test1)[:, 0], np.array(x_test1)[:, 1], 'ro')
plt.plot(np.array(x_test2)[:, 0], np.array(x_test2)[:, 1], 'gx')
plt.plot(np.array(x_test3)[:, 0], np.array(x_test3)[:, 1], 'b.')
# ç”»åˆ†ç±»ç›´çº¿
plt.plot([0, 2], [-w[0][2] / w[0][1], -(w[0][2] + 2 * w[0][0]) / w[0][1]], 'r')
plt.plot([0, 2], [-w[1][2] / w[1][1], -(w[1][2] + 2 * w[1][0]) / w[1][1]], 'g')
plt.plot([0, 2], [-w[2][2] / w[2][1], -(w[2][2] + 2 * w[2][0]) / w[2][1]], 'b')

plt.show()
```
