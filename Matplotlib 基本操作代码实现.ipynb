```python
# 方式二 用subplot2grid

 
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100) 
plt.figure(figsize=(12,9))

ax1 = plt.subplot2grid(shape = (3,3),
                       loc = (0,0),
                      colspan = 3 )

ax1.plot(x,np.sin(10*x))

ax1.set_title('ax1_title')


ax2 = plt.subplot2grid((3,3),(1,0),colspan = 2)
ax2.set_facecolor('green')
ax2.plot(x,np.cos(x),color = 'red')
ax2.set_title('ax2')

ax3 = plt.subplot2grid((3,3),(2,1),colspan = 2)
ax3.plot(x,np.cos(x*2))
ax3.set_title('ax3')





```

```python
# 方式三

 
import numpy as np
import matplotlib.pyplot as plt
# 需要导入gridspec模块
import matplotlib.gridspec as gridspec


x = np.linspace(0,2*np.pi,200)
fig = plt.figure(figsize=(12,9))

gs  = gridspec.GridSpec(3,3) # 切成 3x3 

ax1 = fig.add_subplot(gs[0,:]) # 把切好的用来赋值
ax1.plot(x,np.sin(10*x))
ax1.set_title('ax1.title')



ax2 = fig.add_subplot(gs[1,1:])
ax2.plot(x,np.sin(10*x))



ax3 = fig.add_subplot(gs[-1,0])
ax3.plot(x,np.sin(10*x))
```

```python
import matplotlib.pyplot as plt 
import numpy

loc = np.random.randint(0,10,size = (10,2)) 
plt.figure(figsize=(10, 10))

plt.plot(loc[:,0],loc[:,1],'g*',ms = 40)  # ms是调整点的大小用的
plt.grid(True)


way = np.arange(10)
np.random.shuffle(way)

for i in range(0,len(way)-1) : 
    start = loc[way[i]] # 这个是上边定义的array 不是pd的loc
    end = loc[way[i+1]]
    plt.arrow(start[0],start[1],end[0]-start[0],end[1]-start[1], # 思路是找到起始点，然后用下一个点坐标计算差值，利用勾股定理算长度
             head_width = 0.2 ,lw = 2, # 设置箭头大小和线粗细
             length_includes_head = True)#  长度包含箭头
    
    plt.text(start[0],start[1],s = i ,fontsize = 18 ,color = 'red') # s 是文本内容 必须有的
    

    if i == len(way) -2 : 
        plt.text(end[0],end[1],s = i + 1,fontsize =18,color = 'blue')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def annotate_con_style(ax,connectionstyle):
    x1,y1 = 3,2
    x2,y2 = 8,6
    ax.plot([x1,x2],[y1,y2],'.')
    ax.annotate(s = '' ,
                xy = (x1,y1),
                xytext = (x2,y2),
                arrowprops = dict(arrowstyle = '->',
                                  color = 'red',
                                  shrinkA = 5, shrinkB = 5,
                                  connectionstyle = connectionstyle))
                                  
                                                                         
                                                                    
    ax.text(0.05,0.95,connectionstyle.replace(",",'\n'),
           transform = ax.transAxes,
           ha = 'left',va = 'top')
    
    
```

```python
fig,axs = plt.subplots(3,5,figsize = (18,12))

annotate_con_style(axs[0, 0], "angle3,angleA = 90,angleB =0")
annotate_con_style(axs[0, 1], "angle3,angleA = 0,angleB = 90")
annotate_con_style(axs[0, 2], "angle3,angleA = 90,angleB = 30")
annotate_con_style(axs[0, 3], "angle,angleA = -90,angleB = 180,rad = 5") # 圆角
annotate_con_style(axs[0, 4], "angle,angleA = 0,angleB = 90")

annotate_con_style(axs[1, 0], "arc3,rad = 0")
annotate_con_style(axs[1, 1], "arc3,rad = -0.5")
annotate_con_style(axs[1, 2], "arc3,rad = 0.5")

annotate_con_style(axs[1, 3], "angle,angleA = -90 ,angleB = 10 ,rad = 5")
annotate_con_style(axs[1, 4], "angle,angleA = 60,angleB = 90 ,rad = 7")

annotate_con_style(axs[2, 0], "arc,angleA = -90,angleB =0 ,armA =30 ,armB =30 ,rad =0")
annotate_con_style(axs[2, 1], "arc,angleA = 0,armA =30 ,armB =30 ,rad =0")

annotate_con_style(axs[2, 2], "bar,fraction = 0.3")
annotate_con_style(axs[2, 3], "bar,fraction = -0.3")
annotate_con_style(axs[2, 4], "bar,angle = 180,fraction = -0.2")


# 设置区域刻度 这样就不会有线跑出外边了
for ax in axs.flat:
    ax.set(xlim = (0,10),ylim = (0,10),xticks = [] ,yticks = [] ,aspect = 1)
fig.tight_layout(pad = 0.2)
```

```python
# 甜甜圈

 
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

recipe = ["225g flour",
"90g sugar",
"1 egg",
"60g butter",
"100ml milk", "1/2package of yeast"]

data = [225, 90, 50, 60, 100, 5]

wedges,texts = plt.pie(data,startangle = 40 ,wedgeprops = dict(linewidth = 3 , width = 0.4 , edgecolor = 'w')) # 做饼 

bbox_props = dict(boxstyle = 'square,pad =0.3',fc = 'w', ec ='k',lw = 0.72)

kw = dict(arrowprops = dict(arrowstyle = '-'),
         bbox = bbox_props ,va = 'center')


# 这一大堆都是为了画标签
for i, p in enumerate(wedges) : 
    ang = (p.theta2 - p.theta1)/ 2. + p.theta1
    
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    
    ha = {-1 : 'right',1:'left'}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang) # 箭头连接样式 
    kw["arrowprops"].update({"connectionstyle": connectionstyle}) # 更新箭头连接方式
    
    plt.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y), ha=ha,**kw,fontsize = 18,weight = 'bold')

    
plt.title("Matplotlib bakery: A donut",fontsize = 18)
plt.tight_layout()
 
```
